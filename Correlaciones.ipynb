```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos de tu archivo de Excel en un DataFrame de pandas
data = pd.read_excel('/content/Hogares_11.xlsx')

# Seleccionar la columna del eje x y las dos columnas del eje y
x_col = 'Entidad federativa'
y_cols = ['Emigrante 2020', 'Disponen de energía eléctrica 2020']

# Calcular correlación
corr_matrix = data[[x_col] + y_cols].corr()

# Crear DataFrame con valores de correlación
corr_df = pd.DataFrame(corr_matrix.stack(), columns=["correlation"]).reset_index()
corr_df = corr_df.rename(columns={"level_1": "y_col", "nombre_de_la_columna_x": "x_col"})


# Crear mapa de calor
heatmap = sns.heatmap(corr_df.pivot("y_col", "x_col", "correlation"), cmap="coolwarm", annot=True, fmt=".2f")
heatmap.set_title("Mapa de calor de correlación")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de las dos columnas en un DataFrame
df = pd.read_excel('/content/Hogares_11.xlsx', usecols=['Emigrante 2020', 'Disponen de energía eléctrica 2020'])

# Calcular la correlación entre las dos columnas
corr = df['Emigrante 2020'].corr(df['Disponen de energía eléctrica 2020'])

# Crear un mapa de calor
ax = sns.heatmap(df.corr(), annot=True, cmap='coolwarm')

# Añadir etiquetas de eje y título
ax.set_xlabel('Emigrante 2020')
ax.set_ylabel('Disponen de energía eléctrica 2020')
ax.set_title('Mapa de calor de correlación\nCorrelación: {:.2f}'.format(corr))

# Mostrar el mapa de calor
plt.show()
```
